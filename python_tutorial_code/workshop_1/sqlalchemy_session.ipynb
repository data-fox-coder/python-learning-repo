```python
from sqlalchemy import (create_engine, inspect, text, select, MetaData, Table, and_, or_, desc, asc, func,)

engine_903 = create_engine("sqlite+pysqlite:////workspaces/python-learning-repo/python_tutorial_code/workshop_1/903_database.db")

connection_903 = engine_903.connect()
inspection_903 = inspect(connection_903)

inspection_903.get_table_names()

```

```python
gravity_engine = create_engine("sqlite+pysqlite:////workspaces/python-learning-repo/python_tutorial_code/workshop_1/gravity.db")

gravity_connection = gravity_engine.connect()
gravity_inspection = inspect(gravity_connection)

gravity_inspection.get_table_names()
```

```python
metadata_903 = MetaData()

episodes = Table("episodes", metadata_903, autoload_with=engine_903)
print(episodes.name)   
print(episodes.c.keys())
print(repr(episodes))
```

```python
metadata_gravity = MetaData()

books = Table("book", metadata_gravity, autoload_with=gravity_engine)
print(books.name)   
print(books.c.keys())
print(repr(books))
```

```python
with engine_903.connect() as conn:
    result = conn.execute(text("SELECT * FROM episodes"))
    rslut_proxy = result.fetchall()

first_row = rslut_proxy[0]
print(first_row)
print(first_row.CHILD)
```

```python
with gravity_engine.connect() as conn:
    stmt = "SELECT * FROM book"
    result_proxy = conn.execute(text(stmt))
    result = result_proxy.fetchall()

selectrows = result[1]
print(selectrows)
print(selectrows.title)
```

```python
with gravity_engine.connect() as conn:
    stmt = "SELECT * FROM book"
    result_proxy = conn.execute(text(stmt))
    result = result_proxy.fetchmany(size=10)

print(result)
```

```python
with gravity_engine.connect() as conn:
    stmt = select(books)
    result = conn.execute(stmt).fetchmany(size=5)

print(result)
```

```python
header = Table('header', metadata_903, autoload_with=engine_903)

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU",
                                    header.columns.ETHNIC == "NOBT"))
    result = con.execute(stmt).fetchmany(size=10)
    print(result)
```

```python
books_find = Table('book', metadata_gravity, autoload_with=gravity_engine)

with gravity_engine.connect() as con:
    stmt = select(books_find).where(or_(books_find.columns.book_id == 1,
                                    books_find.columns.book_id == 2))
    result = con.execute(stmt).fetchmany(size=10)
    print(result)
```

```python
import pandas as pd

with gravity_engine.connect() as con:
    stmt = select(books_find)
    result = con.execute(stmt).fetchall()

df = pd.DataFrame(result)
print(df)
```
